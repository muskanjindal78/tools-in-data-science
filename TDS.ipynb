```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data = pd.read_excel("/content/drive/MyDrive/TDS/data/TDS_Week_5_Dataset_1.xlsx")
data

data['Failure Type'].replace({'No Failure':0,'Heat Dissipation Failure':0,'Overstrain Failure':0,'Tool Wear Failure':0,'Random Failures': 1, 'Power Failure' : 0}, inplace=True)

data['Failure Type'].value_counts()
```

```python
Q1 = np.percentile(data['Tool wear [min]'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(data['Tool wear [min]'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

upper = Q3 +1.5*IQR

lower = Q1 - 1.5*IQR

print("Old Shape: ", data.shape)

# Upper bound
upper = np.where(data['Tool wear [min]'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data['Tool wear [min]'] <= (Q1-1.5*IQR))

''' Removing the Outliers '''
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)

print("New Shape: ", data.shape)
```

```python
conditions = [
    (books["average_rating"] < 0.5),
    (books["average_rating"] >= 0.5) & (books["average_rating"] < 1.5),
    (books["average_rating"] >= 1.5) & (books["average_rating"] < 2.5),
    (books["average_rating"] >= 2.5) & (books["average_rating"] < 3.5),
    (books["average_rating"] >= 3.5) & (books["average_rating"] < 4.5),
    (books["average_rating"] >= 4.5) & (books["average_rating"] < 5.5),
]

# create a list of the values we want to assign for each condition
values = [0, 1, 2, 3, 4, 5]

books["bucket"] = np.select(conditions, values)


q16_df = books["bucket"].value_counts()
q16_df = q16_df.reset_index()

q16_df.rename(columns={"index": "Bucket", "bucket": "Count"}, inplace=True)


print(q16_df)


# *************If you are missing any buckets add them manullay Comment/Uncommnet the below code***************


missing_df = pd.DataFrame({'Bucket':[0,1,2],'Count':[0,0,0]}) # Add missing buckets
q16_df = pd.concat([q16_df,missing_df])

q16_df.sort_values(by=['Bucket'])


# Sorting by bucket
q16_df.sort_values(by=["Bucket"], inplace=True)


# write to a csv file , upload it to flourish

q16_df.to_csv("/content/drive/MyDrive/TDS/data/q16_data.csv")

```
